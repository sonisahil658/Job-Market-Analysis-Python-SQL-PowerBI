```python
df.drop(columns=['FIELD4'], inplace = True)
```

```python
df['Company_Name'] = df['Company_Name'].str.split('\n').str[0]
```

```python
df.head()
```

```python
df['Salary_Estimate'] = df['Salary_Estimate'].str.replace(r'\(.*\)', '', regex=True).str.strip()
```

```python
df.head()
```

```python
df.replace(-1, pd.NA, inplace=True)
```

```python
df.head()
```

```python
df['Rating'] = df['Rating'].fillna(df['Rating'].mean())
```

```python
df['Rating']
```

```python
df['Founded'] = df['Founded'].fillna(df['Founded'].median())    # median is better than mean for years

df['Age'] = df['Age'].fillna(df['Age'].median())            # same for age
```

```python
df['Founded']
```

```python
df['Age']
```

```python
df['Lower_Salary'] = df['Lower_Salary'].fillna(df['Lower_Salary'].median())

df['Upper_Salary'] = df['Upper_Salary'].fillna(df['Upper_Salary'].median())

df['Avg_SalaryK'] = df['Avg_SalaryK'].fillna(df['Avg_SalaryK'].median())
```

```python
df['Lower_Salary']
```

```python
df['Upper_Salary']
```

```python
df['Avg_SalaryK']
```

```python
text_cols = ['Size', 'Industry', 'Sector', 'Revenue', 'Competitors', 'Headquarters', 'Type_of_ownership']

df[text_cols] = df[text_cols].fillna('Unknown')    # fill all text NaN with 'Unknown' in one line
```

```python
df.head()
```

```python
df['State'] = df['Location'].str.split(',').str[-1].str.strip()    # "New York, NY" → "NY"
```

```python
df['State'] 
```

```python
df.head()
```

```python
skill_cols = ['Python', 'spark', 'aws', 'excel', 'sql_', 'tableau', 'bi', 'hadoop', 'tensor', 'flink', 'mongo', 'google_an']

df[skill_cols] = df[skill_cols].fillna(0)    # missing skill means not required = 0
```

```python
df[skill_cols]
```

```python
df.isnull().sum()    # should show 0 or very low missing values now

df.shape             # should still be 742 rows

df.head()            # visual check
```

```python
type(df)    # should say <class 'pandas.core.frame.DataFrame'>
```

```python
df.reset_index()    # reset index to clean 0,1,2,3...

df['ID'] = df.index    # assign clean sequential ID to ID column

df['ID'].head()    # verify
```

```python
df['Competitors'] = df['Competitors'].replace('-1', 'Unknown')    # replace text -1 with Unknown
```

```python
df['Competitors']
```

```python
df['Founded'] = df['Founded'].astype(int)    # convert 1973.0 → 1973
```

```python
df['Founded'] 
```

```python
df.to_csv("market_data_clean.csv", index=False)    # overwrite with fully clean file

print("Done!", df.shape)
```
