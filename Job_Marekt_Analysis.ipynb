```python
import nbformat

# Step 1 — Read your notebook file
# Change the filename to your actual notebook name
with open("Job_market.ipynb", "r", encoding="utf-8") as f:
    nb = nbformat.read(f, as_version=4)

# Step 2 — Clean all cell outputs and reset execution count
for cell in nb.cells:
    if cell.cell_type == "code":
        cell.outputs = []
        cell.execution_count = None

# Step 3 — Fix metadata
nb.metadata = {
    "kernelspec": {
        "display_name": "Python 3",
        "language": "python",
        "name": "python3"
    },
    "language_info": {
        "name": "python",
        "version": "3.10.0"
    }
}

# Step 4 — Save clean notebook
with open("Job_market.ipynb", "w", encoding="utf-8") as f:
    nbformat.write(nb, f)

print("Done! Upload Job_market.ipynb to GitHub")
```
